## IQR Method

```python
Q1=wine_testing.quantile(0.25)
Q3=wine_testing.quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

```python
print(wine_testing<(Q1 - 1.5 * IQR))|(wine_testing > (Q3 + 1.5 * IQR))
```
