```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

wine_testing=pd.read_csv('C:\\Users\\assharma\\Downloads\\winequality-red.csv')
wine_testing_df=pd.DataFrame(wine_testing)
wine_testing_df
```

```python
wine_testing_df[wine_testing_df.isnull().values]
```

```python
wine_testing_df=wine_testing_df.dropna(axis=0)
```

```python
target_count=wine_testing.quality.value_counts()
target_count
```

```python
target_count.plot(kind='bar',title='QUALITY COUNT')
```

```python
    import imblearn
    from imblearn.under_sampling import RandomUnderSampler
    x=wine_testing_df.iloc[:,0:12]
    y=wine_testing_df.iloc[:,-1]
    type(y)
```

```python
rus= RandomUnderSampler(return_indices=True)
x_rus,y_rus,id_rus=rus.fit_sample(x,y)
print('Removed indexes:', id_rus)
```

```python
x_rus_template=pd.DataFrame(x_rus)
x_rus_template['QUALITY']=y_rus
```

```python
target_under_sampling=x_rus_template['QUALITY'].value_counts()
target_under_sampling.plot(kind='bar')
```

```python
from imblearn.over_sampling import RandomOverSampler
x=wine_testing_df.iloc[:,0:12]
y=wine_testing_df.iloc[:,-1]
```

```python
ros=RandomOverSampler()
x_ros,y_ros=ros.fit_sample(x,y)
np.size(x_ros)
```

```python
x_ros_template=pd.DataFrame(x_ros)
x_ros_template['QUALITY']=y_ros
x_ros_template
```

```python
target_count_quality=x_ros_template['QUALITY'].value_counts()
target_count_quality.plot(kind='bar',title='QUALITY COUNT')
```

```python
from imblearn.under_sampling import TomekLinks

x=wine_testing_df.iloc[:,1:12]
y=wine_testing_df.iloc[:,-1]

t1=TomekLinks(return_indices=True, ratio='majority')
x_t1,y_t1,i_t1=t1.fit_sample(x,y)
print('Removed Indxes : ',i_t1)
```

```python
x_ros_tom=pd.DataFrame(x_t1)
x_ros_tom['QUALITY']=y_t1
x_ros_tom
```

```python
target_count_tom=x_ros_tom['QUALITY'].value_counts()
target_count_tom.plot(kind='bar',title='QUALITY COUNT')
```
