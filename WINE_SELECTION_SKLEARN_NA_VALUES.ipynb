```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import Imputer

wine_testing=pd.read_csv('C:\\Users\\assharma\\Downloads\\winequality-red.csv')
wine_testing
```

```python
wine_testing.columns=map(str.upper,wine_testing.columns)
wine_testing['DUMMY']=wine_testing['DUMMY'].str.upper()
wine_testing.rename(columns={'FREE SULFUR DIOXIDE':'FRF_SULF_DIX','TOTAL SULFUR DIOXIDE':'TOT_SULF_DIX'},inplace=True)

wine_testing
```

```python
wine_testing.drop(['DUMMY'],axis=1)
```

```python
wine_testing[wine_testing.isnull().values]
```

```python
wine_testing_imputed=wine_testing
imp = Imputer(missing_values='NaN',strategy='mean',axis=0)
```

```python
wine_testing_imputed
```

```python
imp.fit(wine_testing_imputed.drop(['DUMMY'],axis=1))
```

```python
wine_testing_imputed= imp.transform(wine_testing_imputed.drop(['DUMMY'],axis=1))
```

```python
wine_testing_imputed_PD=pd.DataFrame(wine_testing_imputed)
```
